```python
# Import libraries 
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create the vector of positive whole numbers in a list and then change the list to an array. 
# Since we should have n(j,j) pairs then both radius and angle would be the same 
# I could have created one of them only, but I will choose to create two with same output

radius = list(range(0, 1000))
radius = np.asarray(radius)
angle = list(range(0, 1000))
angle = np.asarray(angle)
```

```python
# Now we plot the required spiral with adding label, and visibility feature as well as limits for quarters preview
fig = plt.figure(figsize=(15, 7))
ax = fig.add_subplot(projection='polar')
c = ax.scatter(angle, radius, alpha=0.75, label="Archimedes Spiral")


ax.set_thetamin(0)
ax.set_thetamax(90)
ax.legend()
```

```python
# Now to draw a smooth line through scatter, I will have to define a new array for the line because the command will be different for both as shown below 
radius = list(range(0, 25))
np.asarray(radius)
angle = list(range(0, 25))
np.asarray(angle)
fig = plt.figure()
ax = fig.add_subplot(projection='polar')
c = ax.scatter(angle, radius, alpha=0.75, label="Archimedes Spiral")
ax.plot(np.arange(0,25,0.1),np.arange(0,25,0.1))
```
